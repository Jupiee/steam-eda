# Analysis of Steam's Best Selling Games of All Time (June 1, 2025)

This dataset is a composite, created by merging data from three distinct sources to provide a richer profile for each game. The columns are sourced as follows:

* From Steam: The core data for each game, including its name, review scores, and technical details, was scraped directly from the official Steam store page.
-- `game_name`, `reviews_like_rate`, `all_reviews_number`, `release_date`, `developer`, `user_defined_tags`, `supported_os`, `supported_languages`, `price`, `other_features`.
  
* From GameFAQs: To enrich the dataset with qualitative metrics, data regarding content and player experience was sourced from GameFAQs.
-- `age_restriction`, `rating`, `difficulty`, `length`
-- **Note:** In cases where data for a specific game was unavailable on GameFAQs, these fields were estimated by manually analyzing user reviews and discussions on the game's official Steam page.

* From SteamDB: To provide an estimate of reach and ownership, download figures were sourced from the third-party statistics site SteamDB.
-- `estimated_downloads`
-- **Important Exclusion Criterion:** Only games for which an estimated_downloads value could be obtained from SteamDB were included in the final dataset. Games lacking this data were excluded from the list, even if they appeared in the 'Bestsellers' section.

### Column Descriptions
* `game_name`: The official title of the game.
* `reviews_like_rate`: The recommendation rate from user reviews on Steam (e.g., '95% of the 100 reviews are positive').
* `all_reviews_number`: The total number of user reviews the game has received on Steam.
* `release_date`: The official release date of the game on the Steam platform. This includes the initial release date for games in Early Access.
* `developer`: The primary developer or studio behind the game.
* `user_defined_tags`: A set of genres and categories assigned to the game by the Steam community (e.g., 'RPG', 'Open World', 'Story Rich').
* `supported_os`: A list of the operating systems the game officially supports (win, mac, linux).
* `supported_languages`: A list of languages that the game supports for interface, audio, or subtitles.
* `price`: The price of the game. Note: The currency is MENA - U.S. Dollar, a regional price for the Middle East & North Africa, not the standard USD. A value of 0 in this column indicates the game is 'Free to Play'.
* `other_features`: A set of features defined by Steam under the 'player support' category (e.g., 'Single-player', 'Online PvP', 'Steam Achievements', 'Family Sharing', 'Online Co-op', 'VR Supported').
* `age_restriction`: The recommended age restriction for the game's content, encoded as follows: 0 (Everyone), 10 (10+), 13 (13+), 17 (17+).
* `rating`: An overall user-provided rating for the game on a scale of 1 to 5, where 1 is the lowest and 5 is the highest.
* `difficulty`: An estimated difficulty of the game as perceived by players, on a scale of 1 to 5, where 1 is the easiest and 5 is the hardest.
* `length`: The average time (in hours) players spend to complete or fully experience the game. For this dataset, the value is capped at a maximum of 80 hours.
* `estimated_downloads`: The estimated total number of owners for the game.
